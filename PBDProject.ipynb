```python
sample="""
Ms. Marvel (TV series)
From Wikipedia, the free encyclopedia
Jump to navigationJump to search
Ms. Marvel
Ms. Marvel (TV series) logo.jpeg
Genre	
Action-adventure
Comedy
Coming-of-age
Superhero
Created by	Bisha K. Ali
Based on	Marvel Comics
Starring	
Iman Vellani
Matt Lintz
Yasmeen Fletcher
Zenobia Shroff
Mohan Kapur
Saagar Shaikh
Laurel Marsden
Azhar Usman
Rish Shah
Arian Moayed
Alysia Reiner
Laith Nakli
Nimra Bucha
Travina Springer
Adaku Ononogbo
Samina Ahmad
Fawad Khan
Mehwish Hayat
Farhan Akhtar
Aramis Knight
Music by	Laura Karpman
Country of origin	United States
Original language	English
No. of episodes	4
Production
Executive producers	
Kevin Feige
Louis D'Esposito
Victoria Alonso
Brad Winderbaum
Sana Amanat
Adil & Bilall
Bisha K. Ali
Production locations	
Atlanta, Georgia
New Jersey
Thailand
Cinematography	
Robrecht Heyvaert
Carmen Cabana
Jules O'Loughlin
Editors	
Nona Khodai
Sabrina Plisco
Emma McCleave
Sushila Love
Running time	48–52 minutes
Production company	Marvel Studios
Distributor	Disney Platform Distribution
Release
Original network	Disney+
Original release	June 8, 2022 –
present
Chronology
Related shows	Marvel Cinematic Universe television series
Ms. Marvel is an American television miniseries created by Bisha K. Ali for the streaming service Disney+, based on the Marvel Comics featuring the character Kamala Khan / Ms. Marvel. It is the seventh television series in the Marvel Cinematic Universe (MCU) produced by Marvel Studios, sharing continuity with the films of the franchise. It follows Kamala Khan, a 16-year-old fangirl of the Avengers, who struggles to fit in until she gains her own powers. Ali serves as head writer with Adil & Bilall (Adil El Arbi and Bilall Fallah) leading the directing team.

Iman Vellani stars as Kamala Khan / Ms. Marvel, with Matt Lintz, Yasmeen Fletcher, Zenobia Shroff, Mohan Kapur, Saagar Shaikh, Laurel Marsden, Azhar Usman, Rish Shah, Arian Moayed, Alysia Reiner, Laith Nakli, Nimra Bucha, Travina Springer, Adaku Ononogbo, Samina Ahmad, Fawad Khan, Mehwish Hayat, Farhan Akhtar, and Aramis Knight also starring. The series was announced with Ali's involvement in August 2019. Vellani was cast in September 2020, with El Arbi and Fallah, Meera Menon, and Sharmeen Obaid-Chinoy hired as the series' directors. Filming began in early November 2020, shooting in Atlanta, Georgia, and New Jersey, before concluding in Thailand in May 2021.

Ms. Marvel premiered on June 8, 2022, and will consist of six episodes, concluding on July 13. It is part of Phase Four of the MCU. The series will serve as set-up for the film The Marvels (2023), in which Vellani will reprise her role as Khan along with additional cast members from the series. The series recieved positive reviews, particularly for Vellani's performance.


Contents
1	Premise
2	Cast and characters
3	Episodes
4	Production
4.1	Development
4.2	Writing
4.3	Casting
4.4	Design
4.5	Filming
4.6	Post-production
4.7	Music
5	Marketing
6	Release
7	Reception
8	Future
9	References
10	External links
Premise
Kamala Khan, a 16-year-old fangirl of the Avengers, particularly Carol Danvers / Captain Marvel, struggles to fit in until she gains her own powers.[1]

Cast and characters
Iman Vellani as Kamala Khan / Ms. Marvel:
A 16-year-old Pakistani-American high school student from Jersey City who is an aspiring artist and avid gamer, and writes superhero fan fiction about heroes such as Captain Marvel.[1][2] Khan gains the ability to harness cosmic energy and create hard light constructs from a magical bangle.[3][4] Executive producer Sana Amanat said Khan brought a "grounded perspective" to the MCU with "bright and eager eyes". Vellani noted Khan enjoys the simplicity of a life with powers, instead of worrying about high school, boys and relationships, family drama, and culture and religion.[5] She also added that the character "felt so much like me" due to their similar background, ethnicity, and love for the Marvel universe.[6] Head writer Bisha K. Ali described Khan as an "avatar" for all of the viewers who had grown up with the MCU.[7]
Matt Lintz as Bruno Carrelli: Kamala's best friend who has a crush on her.[8][9] Lintz called Carrelli "very loyal" to Kamala, more so after she gets her powers, as it creates a new dynamic between them.[10]: 7 
Yasmeen Fletcher as Nakia Bahadir:
Kamala's close friend.[11] Fletcher called Bahadir a "strong character" who is "opinionated and is willing to fight for anything that she believes in", as well as a character that "breaks a lot of the stereotypes for hijab girls".[10]: 7–8 
Zenobia Shroff as Muneeba Khan:
Kamala's mother and Yusuf's wife.[12] She is stricter on Kamala than Yusuf, with Shroff saying it comes from her knowing Kamala's potential while still wanting to "protect her from all that".[10]: 8 
Mohan Kapur as Yusuf Khan:
Kamala's father and Muneeba's husband.[12] Kapur described Yusuf as "a warm, caring, compassionate man", who is more understanding of Kamala and her dreams than Muneeba.[10]: 8 
Saagar Shaikh as Aamir Khan:
Kamala's older brother.[13] Shaikh described Aamir as "kind of aloof and is not as funny as he thinks he is", who is very religious and serves as a middleman between Kamala and their parents.[10]: 8–9 
Laurel Marsden as Zoe Zimmer: The most popular girl at Kamala's high school and a social media influencer.[13]
Azhar Usman as Najaf:[10]: 15  A halal food vendor acquainted with Kamala.
Rish Shah as Kamran:
A boy Kamala has a crush on.[12] Shah believed Kamran is able to relate to Kamala so easily because he had a "a lack of belonging and community" and is able to express himself culturally around her.[10]: 9 
Arian Moayed as P. Cleary: A Department of Damage Control (DODC) agent.[14]
Alysia Reiner as Sadie Deever: A DODC agent working with Cleary.[14]
Laith Nakli as Sheikh Abdullah: An imam at Kamala's mosque.[13]
Nimra Bucha as Najma: Kamran's mother and a Djinn, who is the leader of the Clandestines trying to return to their home Noor dimension after being exiled on Earth.[13]
Travina Springer as Tyesha Hillman: Aamir's wife.[13]
Adaku Ononogbo as Fariha: A member of the Clandestine who wields a spear.[13]
Samina Ahmad as Sana: Kamala's grandmother who lives in Karachi, Pakistan.[15] Zion Usman portrays a young Sana.[citation needed]
Fawad Khan as Hasan:[16] Kamala's great-grandfather.
Mehwish Hayat as Aisha: Kamala's great-grandmother, and the original owner of the bangle.[17]
Farhan Akhtar as Waleed: The leader of the Red Daggers, a group of vigilantes who wear red bandanas and wield throwing knives.[15]
Aramis Knight as Kareem / Red Dagger: A member of the Red Daggers.[15]
Recurring in the series are Anjali Bhimani and Sophia Mahmud as Aunties Ruby and Zara.[18][citation needed] Additional guest stars include Jordan Firstman as school counselor Gabe Wilson;[19][13] Ali Alsaleh and Dan Carter as Aadam and Saleem, members of the Clandestine who wield a golden mace and powered whip, respectively;[13] Vardah Aziz and Asfandyar Khan as Kamala's cousins Zainab and Owais, respectively;[15] and Alyy Khan in an undisclosed role.[20] Ryan Penagos, vice president and creative executive at Marvel New Media, has a cameo appearance as the cosplay competition host at AvengerCon.[21] Directors Adil El Arbi and Bilall Fallah make cameo appearances as a man in the mosque and one of the "Mosque Bros", respectively,[citation needed] while Amanat cameos during Aamir and Hillman's wedding.[22]

Episodes
No.	Title	Directed by [23]	Written by [24]	Original release date [25]
1	"Generation Why"	Adil & Bilall	Bisha K. Ali	June 8, 2022
Kamala Khan is a 16-year-old high schooler and fangirl of the Avengers, particularly Captain Marvel. After failing another driving test and meeting with school counselor Gabe Wilson, Kamala and her best friend Bruno Carrelli finish her Captain Marvel cosplay for "AvengerCon" while avoiding her strict parents, Yusuf and Muneeba. Kamala receives a package of items from her grandmother Sana, including a golden bangle, but Muneeba sees it as junk. After failing to convince her parents to let her go to AvengerCon, Kamala sneaks out with Bruno to attend anyway, taking the bangle as part of her cosplay. After getting there and dressing up, she puts it on, but it causes her to project constructs of cosmic energy that inadvertently cause havoc. Amidst the chaos, Kamala uses her powers to save her classmate Zoe Zimmer. Bruno rushes Kamala home, where a distraught Muneeba pleads with her to focus on her own story. In a mid-credits scene, Department of Damage Control (DODC) agents P. Cleary and Sadie Deever watch a video of Kamala's incident at AvengerCon and head to New Jersey to find and detain her.
2	"Crushed"	Meera Menon	Kate Gritmon	June 15, 2022
Kamala begins training to control her powers with Bruno, who deduces that the bangle activated her own superpowers. Kamala, Bruno, and Nakia Bahadir attend a party organized by Zimmer, where they meet new senior Kamran. Kamala befriends Kamran, but Bruno becomes frustrated when she chooses to spend time with Kamran instead of training. After seeing a vision of a mysterious woman, Kamala asks Sana and Muneeba about her great-grandmother Aisha, the bangle's original owner, but both dismiss her. Yusuf says young Sana lost her way during the partition of India, but was mysteriously able to find her father again. After questioning Zimmer about her savior at AvengerCon, Cleary and Deever order a sweep of the tri-state area, targeting South Asian communities. At an Eid celebration, a young boy slips from a mosque balcony and almost falls before Kamala saves him with her powers. Kamala is chased by DODC agents led by Deever but is saved by Kamran, who introduces her to his mother Najma, the woman from Kamala's vision.
3	"Destined"	Meera Menon	Teleplay by : Freddy Syborn and A. C. Bradley & Matthew Chauncey
Story by : Freddy Syborn	June 22, 2022
Najma explains that she and Kamran are part of a group of five enhanced beings known as Clandestines or Djinn who were exiled from the Noor dimension, and that Aisha was also one of them. She also reveals that the bangle might be able to help them return, and asks for Kamala's help. Kamala agrees, but Bruno warns her that interdimensional travel could be dangerous, so she asks Kamran for more time to ensure that they can do it safely. Kamran assents, but an impatient Najma decides to force Kamala to help them. Kamala's brother Aamir marries his fiancée Tyesha, but Kamran crashes the wedding to warn Kamala before the other Clandestines arrive. Kamala, Bruno, and Kamran are overpowered by the Clandestines while Najma tries to use the bangle, which triggers a vision of a train. The DODC agents arrive and capture the Clandestines, including Kamran. As Kamala and Bruno escape, Nakia sees Kamala using her powers. Sana contacts Kamala, revealing that she also saw the vision of the train and insisting that Kamala and Muneeba must visit her in Karachi.
4	"Seeing Red"	Sharmeen Obaid-Chinoy	Teleplay by : Sabir Pirzada and A. C. Bradley & Matthew Chauncey
Story by : Sabir Pirzada	June 29, 2022
Kamala and Muneeba travel to Karachi and reunite with Sana, who later reveals to Kamala that the bangle is trying to convey a message through the vision of the train. The next day, a masked Kamala goes to the train station to investigate, but is attacked by Kareem, a member of the Red Daggers vigilante group, who initially mistakes her for one of the Clandestines. Kareem takes Kamala to the Red Daggers' hideout, where Kamala learns from their leader Waleed that the Clandestines are trying to break the Veil of Noor, which separates the Djinns' dimension from the human world, in order to expand and take over. The Clandestines escape the DODC's supermax prison, but Najma abandons Kamran for helping Kamala. Kamala begins training with the Red Daggers to master her powers, but they are interrupted by the Clandestines. A chase ensues, during which Waleed kills one of the Clandestines but is fatally stabbed by Najma. As Kamala and Kareem fend off the Clandestines, Kareem kills one of them and Najma accidentally stabs the bangle, sending Kamala into a vision of the partition.
5	TBA	Sharmeen Obaid-Chinoy	Fatimah Asghar	July 6, 2022
6	TBA	Adil & Bilall	Teleplay by : Will Dunn and A. C. Bradley & Matthew Chauncey
Story by : Will Dunn	July 13, 2022
Production
Development
Marvel Entertainment's Creative Consultant Joe Quesada said in September 2016 that there were plans to explore the character of Kamala Khan / Ms. Marvel in other media following her unusually rapid success and popularity among comic book readers.[26] Marvel Studios President Kevin Feige said in May 2018 that a Marvel Cinematic Universe (MCU) project based on Kamala Khan was "in the works",[27][28] and would follow the release of the film Captain Marvel (2019) as Khan is inspired by that film's title character Carol Danvers.[27]

By August 2019, Marvel Studios had begun development on a Ms. Marvel television series for the streaming service Disney+, with Bisha K. Ali hired to serve as the head writer after working as a writer on the series Loki.[29][30] Ali approached Kevin Wright, one of the Marvel Studios executives on Loki, while they were working together to get an opportunity to meet about Ms. Marvel,[30] and declined an extension to continue working on Loki in order to create her pitch. She believed Marvel responded to her "general vibe and energy" from her pitch, and that a series about Ms. Marvel was very personal to her "as a fan, as a South Asian person, as a Pakistani woman, [and] as a woman from a Muslim background", making a show about these be "specific and universal".[31] Ms. Marvel was officially announced at the 2019 D23 conference.[32]

In September 2020, Adil El Arbi and Bilall Fallah (credited as Adil & Bilall[23]) were hired to direct two episodes of the series, with Meera Menon hired to direct one episode, and Sharmeen Obaid-Chinoy hired to direct three episodes;[33][34] Menon and Obaid-Chinoy ultimately directed two episodes each.[23] Despite their different visual styles and prior work, all of the directors worked closely together to create a "seamless transition" between their episodes.[10]: 5  Ali, El Arbi, and Fallah serve as executive producers on the series and worked closely with Marvel Studios to develop it,[35][36] with Kamala Khan co-creator Sana Amanat also serving as an executive producer.[36][23] Marvel Studios' Feige, Louis D'Esposito, Victoria Alonso, and Brad Winderbaum also serve as executive producers.[23]

Writing
In addition to Ali, writers for the series include Kate Gritmon, Freddy Syborn, A. C. Bradley, Matthew Chauncey, Sabir Pirzada, Fatimah Asghar, and Will Dunn.[24] Bradley and Chauncey were respectively the head writer, and story editor and writer for What If...?,[37][38] with Bradley also serving as a consulting producer on Ms. Marvel,[38]: 6  while Pirzada was a writer for Moon Knight.[39] Ali believed having Amanat as an executive producer on the series allowed them to "stay true to the character" from the comic books, while adding "something new, with a freshness, vitality and contemporary edge".[40] Ali called her time working on Loki as "good training ground" to serve as Ms. Marvel's head writer,[30] learning what creative freedoms Marvel Studios allowed for each series to inhabit their own part of the MCU, while still being bounded by the overarching storytelling.[30][31] She was told to keep Khan's story "ground level", which was already part of her plan for the series, while always focusing on who the character of Khan represents. Ali also received advice from WandaVision and The Falcon and the Winter Soldier head writers Jac Schaeffer and Malcolm Spellman.[31]

In Ms. Marvel, Khan gains the ability to harness cosmic energy and create hard light constructs from a magical bangle,[3][4] which differs from the shapeshifting abilities that she has in the comics. Feige explained that the Inhuman source of her abilities in the comics did not "match" with the timeline and events of the MCU, so her powers were adjusted to be related to her Pakistani heritage. They were also brought closer to the cosmic powers of the other heroes in the film The Marvels (2023), in which Vellani co-stars. Feige added that the character's "giant hands and arms" would still appear in the series "in spirit".[41] The change in Khan's powers was not part of Ali's initial pitch, with Ali adding it was a "group" decision to alter them and echoed Feige in saying the change was done to align the character with what has been established in the MCU.[40] She added that it was important that the new portrayal of Khan's powers had the "same connectivity with her psychology and the journey she's going through and the way she sees herself" as her shapeshifting abilities did in the comics.[7] Amanat and another of the character's co-creators, G. Willow Wilson, both were consulted on and supported the change, with Amanat feeling that it was "fun to give Kamala different kinds of powers that feel big in scope and cinematic... the essence of what the powers are in the comics is there, both from a metaphorical standpoint and from a visual standpoint."[28]

Ms. Marvel is a coming-of-age story, with Amanat noting the series took inspiration from the character's comics, creating the MCU version of the comics' "quirkiness and stylistic distinction" and being "told through the lens of Kamala's experience and her wild imagination". She also described Khan's life and world as "naturally colorful" given Jersey City is "a pretty crazy and vibrant and multicultural place". Carol Danvers is Khan's inspiration to become a hero, having seen her almost defeat Thanos during the events of Avengers: Endgame (2019). Amanat believed it was important to tell a story about who your heroes are for a young person of color, and exploring what that does "to your sense of self".[28] Ali wanted to mix the "whimsical and magical" qualities from Ms. Marvel's comics with American high school films, being inspired by John Hughes's films and the films 10 Things I Hate About You (1999), Lady Bird (2017), Eighth Grade (2018), and Booksmart (2019),[5] as well as the MCU film Spider-Man: Homecoming (2017). Bollywood actor Shah Rukh Khan also was an inspiration for the series.[30] The series features a love triangle between Khan, Bruno Carrelli, and Kamran, which was substantially toned down during the writing process because Ali did not want audiences to believe it was the dominant storyline. El Arbi called Bruno being trapped in the friend zone "relatable", while Ali felt this was a "fundamental part of growing up".[9]

Casting
In September 2020, newcomer Iman Vellani was cast in the lead role of Kamala Khan.[2] Her aunt had forwarded her a casting call for the role, which led to Vellani submitting a self-tape, before being asked to audition at Marvel Studios' offices in Los Angeles, California.[6] She had an in-person screen test in February 2020, as well as a virtual one over Zoom in June 2020 before landing the role.[42] Set photos later that November revealed Matt Lintz had been cast in the series.[43] The next month, Marvel confirmed that Lintz had been cast,[44] portraying Carrelli,[8] along with Yasmeen Fletcher as Nakia Bahadir;[11] Zenobia Shroff as Muneeba Khan; Mohan Kapur as Yusuf Khan;[12] Saagar Shaikh as Aamir Khan;[45][13] Azhar Usman as Najaf;[10]: 15  Aramis Knight as Kareem / Red Dagger;[8] Rish Shah as Kamran;[46][47] Travina Springer as Tyesha Hillman; Laith Nakli as Sheikh Abdullah; and Nimra Bucha as Najma.[13]

Newcomer Laurel Marsden was cast in the role of Zoe Zimmer in February 2021,[48] while Alysia Reiner was revealed to have also been cast in the series as Department of Damage Control (DODC) agent Sadie Deever.[49][14] Fawad Khan revealed that he was part of the cast in December 2021,[50] appearing as Hasan,[16] and Mehwish Hayat was reported a month later to also have a role in the series,[51] as Aisha.[17] In March 2022, Adaku Ononogbo was revealed to be cast as Fariha.[52] That May, Arian Moayed was revealed to be cast in the series,[53][23] reprising his role as DODC agent P. Cleary from Spider-Man: No Way Home (2021),[14] as was Farhan Akhtar in the guest role of Waleed[54][15] Samina Ahmad appears as Kamala's grandmother Sana.[15]

Also in February 2021, Alyy Khan was revealed to have also been cast in the series.[20] In March 2022, Anjali Bhimani was revealed as auntie Ruby,[55][18] a recurring role, after previously appearing in the Marvel Television series Runaways.[55] Jordan Firstman appears as Gabe Wilson, named as a homage to G. Willow Wilson,[19] while Ali Alsaleh and Dan Carter appear as Aadam and Saleem,[13] and Vardah Aziz and Asfandyar Khan appear as Kamala's cousins Zainab and Owais, respectively.[15]

Design
Arjun Bhasin serves as the costume designer for the series.[40] Khan's Ms. Marvel costume was inspired by burkinis and shalwar kameez, with subtle "cultural details" in the fabric.[56]: 89  Amanat credited Bhasin for incorporating "cool textures" to make the costume feel South Asian in "a very subtle way", and noted the lightning bolt on the costume was updated from the comic version of the costume.[40] Christopher Glass serves as production designer.[57]

Filming
El Arbi and Fallah, Menon, and Obaid-Chinoy, directed the series.[33] Filming began in early November 2020 at Trilith Studios in Atlanta, Georgia,[58][59] with additional filming at Blackhall Studios and Areu Brothers Studios.[60] The series was filmed under the working title Jersey.[61] Robrecht Heyvaert served as the cinematographer for El Arbi and Fallah,[62] with Carmen Cabana cinematographer for Menon,[63] and Jules O'Loughlin serving as cinematographer for Obaid-Chinoy.[64][10]: 27 

The animated sequences were not scripted,[62] though an early version of the script for the first episode had a moment inside Khan's head, which the directors took and expanded that idea across the entire series.[7] El Arbi and Fallah conceived them as a way "to get inside Kamala Khan's head and capture her dream world and fantasy world",[62] and for viewers to see the world "through her eyes".[56]: 88  These moments were a way to translate the vibrancy the directors enjoyed from the comics into the series, and were inspired by the animated film Spider-Man: Into the Spider-Verse (2018).[65] The integration of the animated elements took a lot of planning during pre-production, and did not allow the directors any ability to improvise how those sequences were shot.[65] El Arbi noted the other directors also enjoyed using them for their episodes.[62] The duo were also inspired by Spike Lee and making New Jersey a character in the series like Lee does with New York City in his films,[65] Steven Spielberg, anime (particularly for the visual effects),[56]: 89  John Hughes's films,[65] the films Booksmart and Scott Pilgrim vs. the World (2010),[56]: 89 [65] the early 1990s television series Parker Lewis Can't Lose and Saved by the Bell, and Adrian Alphona and Jamie McKelvie's art from the Ms. Marvel comics.[65] El Arbi and Fallah concluded their episodes on March 5, 2021.[66]

Cultural advisors were used to examine the scripts and what was being filmed to ensure their authenticity.[63] Secondary filming occurred in Hudson County, New Jersey, from March to early April.[60][67] Vellani and Shah continued to film scenes in Atlanta through late April and early May.[47][68] Obaid-Chinoy began filming the fourth and fifth episodes in Thailand by March 23, specifically in Bangkok and at the Studio Park facilities.[69][64] Given the series' strict COVID-19 protocols, Ms. Marvel was able to obtain a waiver from the Thai government to continue filming in the country throughout April and May 2021 despite new restrictions put in place in April that suspended other film and television productions. The cast and crew of 450 people were divided into three bubbles in order for production to continue if a positive test was found in one of the bubbles. Filming for the series concluded in Thailand in early May 2021.[69] Reshoots, with Obaid-Chinoy directing, occurred at the end of January 2022.[70]

Post-production
Editors on the series include Nona Khodai (episodes one, four, five, and six), Sabrina Plisco (episodes one, three, and six), and Emma McCleave and Sushila Love (episodes two and three).[10]: 28  Visual effects for the series were created by Digital Domain, Framestore, FuseFX, Method Studios, RISE, SSVFX, Trixter, and Perception.[71]

Music
External audio
audio icon Laura Karpman - Ms. Marvel Suite (From "Ms. Marvel"/Audio Only) presents composer Laura Karpman's main theme for the series, YouTube audio-only video from MarvelMusicVevo's channel
In May 2022, Laura Karpman was revealed to have composed the score for the series, after previously composing for What If...? and being hired to score The Marvels. Karpman said she immersed herself in Khan's "rich musical heritage".[72] The series features a number of existing songs, with Ali believing music was "an integral part" of adapting the character to the screen, and credited Amanat for having a strong vision of which songs were needed and should be used. Amanat called Ms. Marvel a "great platform" to showcase South Asian music, with the selections a "blend" much in the way the character of Khan is as well.[40] Karpman's score for the series was released digitally by Marvel Music and Hollywood Records in two volumes: music from the first three episodes was released on June 22, 2022,[73] and music from the last three episodes will be released on July 13. The series' main theme, "Ms. Marvel Suite", was released as a single on June 7.[74]

Ms. Marvel: Vol. 1 (Episodes 1-3) [Original Soundtrack][73]
No.	Title	Length
1.	"Ms. Marvel Suite"	4:30
2.	"Figure Out Your Life"	1:24
3.	"Mysterious Parcel"	3:30
4.	"The Plan"	2:18
5.	"Sure They Do"	2:04
6.	"Siren Call"	1:58
7.	"The Save"	5:20
8.	"Cosmic"	1:35
9.	"About Last Night"	1:30
10.	"Bathroom Confab"	1:02
11.	"Trail of Stars"	2:33
12.	"Damage Control"	1:35
13.	"Hard Landing"	3:22
14.	"Flee the Scene"	2:08
15.	"Digging in the Dirt"	3:26
16.	"Bad Things Happen"	1:48
17.	"Questioning"	0:52
18.	"Good News, Bad News"	1:13
19.	"A Thing You Do"	1:43
20.	"Circle Q Score"	1:19
21.	"Bonding"	1:39
22.	"Not Asking Anymore"	2:53
23.	"Blitz"	1:17
24.	"Showdown"	2:03
25.	"Homecoming"	2:09
Total length:	49:29
Marketing
A first look at the series with early footage, testimonials about the impact the character has had, and clips of Vellani's audition was shown at Disney's Investor Day in December 2020. Josh Weiss at Syfy Wire called it "absolutely priceless" seeing the exact moment Vellani learned she had been cast as Khan.[36] Additional footage was shown during Disney+ Day on November 12, 2021.[75][1]

The first trailer for the series was released on March 15, 2022.[3] Many commentators noted the change to Khan's powers from the comics (in which she has shape-shifting abilities) to being able to harness cosmic energy and form constructs with magical bangles,[3][76][77][78] believed to be similar to the Nega-Bands from the comics.[79] Linda Codega at Gizmodo said the trailer was "as adorable, fun, and exciting as the comics[, full] of absolutely excellent directorial decisions" and that the series looked like "a tone-perfect mashup of Scott Pilgrim vs. The World and Sky High".[80] Gabrielle Sanchez of The A.V. Club said the trailer offered "an illustrative guide to Kamala Khan".[78] The final shot of the trailer, as well as a teaser poster also released for the series, pays homage to the cover of Ms. Marvel (2014) #5.[12][81]

A Fan's Guide to Ms. Marvel, a documentary short featuring an exclusive look at the production of the series and interviews from the filmmaking team and Vellani, was released on Disney+ on June 1, 2022.[82][23] In January 2021, Marvel announced their "Marvel Must Haves" program, which reveals new toys, games, books, apparel, home decor, and other merchandise related to each episode of Ms. Marvel following an episode's release,[83] which started on June 7, 2022.[84] QR codes were included in the first two episodes linking viewers to a website to access free digital comics featuring Ms. Marvel that updated weekly;[85][86] this followed a similar program that happened with Moon Knight.[85] The comics released for the episodes, in order, were Ms. Marvel (2014) #1 and Ms. Marvel (2014) #15.[85][86]

Release
Ms. Marvel debuted on Disney+ on June 8, 2022,[3] and will consist of six episodes,[69] releasing weekly until July 13.[25] The red carpet premiere occurred on June 2 at El Capitan Theatre in Los Angeles.[87] The series was originally scheduled to debut in late 2021,[44] but following the announcement of Hawkeye premiering in late November 2021, Aaron Couch of The Hollywood Reporter said it was "unclear" if Ms. Marvel would also still release in 2021, given Hawkeye would still be releasing into December and it was "unlikely" two Marvel Studios series would release at the same time.[88] In August 2021, Matt Webb Mitovich of TVLine felt it was "very safe to assume" Ms. Marvel would premiere in early 2022,[89] with it confirmed the following month to be moving to 2022.[90] In November 2021, it was confirmed to be releasing in mid-2022,[1] with the June 2022 premiere confirmed in March 2022.[3] Ms. Marvel is part of Phase Four of the MCU.[91]

The series will receive a three-part theatrical release across Pakistan through licenser HKC Entertainment, due to Disney+ not being available in the country at that time. The first two episodes were released on June 16, 2022, and will be followed by the third and fourth episodes on June 30, and the last two episodes on July 14.[92][93]

Reception
Ms. Marvel: Critical reception by episode
Percentage of positive reviews tracked by the website Rotten Tomatoes[94]
The review aggregator website Rotten Tomatoes reported a 98% approval rating, based on 235 reviews. The website's critics consensus reads, "Ms. Marvel is a genuinely fresh addition to the MCU—both stylistically and substantively—with Iman Vellani ably powering proceedings with her super-sized charisma."[94] Metacritic, which uses a weighted average, assigned a score of 78 out of 100 based on 23 critics, indicating "generally favorable reviews".[95]

The first two episodes were given to critics to review the series ahead of its premiere,[96] with Destiny Jackson at Empire feeling the series was able to balance the comedic and dramatic moments, with "witty visual gags and well-rounded, warmly funny, instantly likeable characters", adding that Ms. Marvel had "an authentic take on Pakistani-American culture; [...] the show takes admirable care in explaining the rituals and realities of life as a modern Muslim".[97] Writing for Rolling Stone, Alan Sepinwall commented that the superhero scenes were "rarely more than functional" with a lack of energy compared to the more personal ones such as Kamala's "arguments with her parents, her and Nakia protesting the condition of the gender-segregated parts of the mosque, or her and... Kamran flirting while discussing their favorite Bollywood films".[98] Variety's Caroline Framke felt Ms. Marvel broke through "Marvel's typical house style" with the use of animation throughout to convey Khan's thoughts and day dreams, calling it a "tangibly kinetic magic" and likening it to the animated film Spider-Man: Into the Spider-Verse.[99] Activist Malala Yousafzai praised the series for its reflection of Pakistani people,[100] and was "struck by how familiar Kamala Khan’s life seemed to me", saying the series was "for every young person finding their place in the world".[101]

Vellani's portrayal of Kamala Khan received widespread praise.[102][103][96] IGN's Emma Fraser compared her to Hailee Steinfeld as Kate Bishop in Hawkeye,[102] with Kathryn Porter of Paste adding that Vellani "shines" in the role and that there was "no way to explain how great she is in this other than to say that she embodies the true spirit of Kamala Khan".[103] Mohammad Zaheer at BBC Culture called Vellani "an adorable bundle of charisma" in a role "tailor-made for her", likening her to Robert Downey Jr. as Tony Stark and Emma Watson as Hermione Granger from the Wizarding World franchise.[104]

The changes to Khan's powers from the comics was also highlighted,[105][103][98][106] with several critics commenting on the new origin's more personal connection.[107][99][108] In contrast, Sepinwall called these abilities "more generic" than the comics powers.[98] Charles Pulliam-Moore at The Verge wrote that the live-action abilities "are only able to approximate the flashy aspects of what was originally a nuanced metaphor in the comics. [...] But the show doesn't go nearly as far with its hero in terms of using its conceit to explore ideas like internalized racism or the pressures Western (read: white) beauty standards put on people of color".[106] G. Willow Wilson, one of the character's creators, had previously described that during the development of Khan's powers, it was chosen not to give her "sparkly, hand wave-y, floaty, pretty powers", which Porter felt was "one of the most important things about her in the comics, and losing that in favor of powers that are, in fact, sparkly, hand wave-y, floaty, and pretty is really unfortunate".[103]

After the release of the first episode, the series was review bombed on IMDb with a high amount of 1 star ratings. Multiple reviews were also posted with some reviewers not liking the family friendly aspects of the series or the change of Khan's powers, while others called the series "woke", mentioning Captain Marvel.[109][110]

Future
See also: The Marvels
During the series' official announcement at D23, Feige stated that after introducing Ms. Marvel in the series, the character would cross over to the MCU films,[32] which he reiterated in November 2019.[111] Previously, Captain Marvel star Brie Larson expressed interest in including Ms. Marvel in that film's sequel,[112] and Vellani was confirmed to be appearing in the Captain Marvel sequel, The Marvels, in December 2020.[113][114] In February 2021, Feige revealed that Ms. Marvel would set-up The Marvels.[115] The Marvels was not fully conceived by the time Ali started working on the series, so she had a lot of creative freedom in crafting the series, but knew there needed to be a "maturity" across the arc of the series so Khan would "be ready" for the events of The Marvels.[7] Additionally, Shaikh, Shroff, and Kapur reprise their respective roles as Khan's family members in The Marvels.[116]

References
 Hipes, Patrick (November 12, 2021). "Disney+ Day: All The Streamer's Film & TV News From Premiere Dates To Series Orders". Deadline Hollywood. Archived from the original on November 12, 2021. Retrieved November 12, 2021.
 Kroll, Justin (September 30, 2020). "Newcomer Iman Vellani To Play Title Role In Marvel's 'Ms. Marvel' Series For Disney Plus". Deadline Hollywood. Archived from the original on September 30, 2020. Retrieved September 30, 2020.
 Vary, Adam B. (March 15, 2022). "'Ms. Marvel' Trailer Introduces Marvel Studios' First Muslim Superhero in Disney Plus Series". Variety. Archived from the original on March 15, 2022. Retrieved March 15, 2022.
 Wasserman, Ben (June 16, 2022). "Ms. Marvel Fan Catches An Extremely Subtle Easter Egg from an Iconic Comic Cover". Comic Book Resources. Archived from the original on June 18, 2022. Retrieved June 18, 2022.
 Truitt, Brian (May 19, 2022). "Exclusive sneak peek: Muslim teen 'Ms. Marvel' brings new perspective in Disney+ series". USA Today. Archived from the original on May 25, 2022. Retrieved May 30, 2022.
 Sun, Rebecca (June 1, 2022). "'Ms. Marvel' Star Iman Vellani on Playing Kamala Khan: "She Felt So Much Like Me"". The Hollywood Reporter. Archived from the original on June 5, 2022. Retrieved June 5, 2022.
 Coggan, Devan (June 16, 2022). "Ms. Marvel writer Bisha K. Ali on the 'responsibility' of bringing Kamala Khan to the screen". Entertainment Weekly. Archived from the original on June 18, 2022. Retrieved June 18, 2022.
 Abdulbaki, Mae (December 14, 2020). "Ms. Marvel Disney Plus Release Date, Trailer, Cast of the MCU Show". Inverse. Archived from the original on December 14, 2020. Retrieved December 14, 2020.
 Paige, Rachel (June 22, 2022). "'Ms. Marvel': Exploring Kamala Khan and Bruno Carrelli's Relationship". Marvel.com. Archived from the original on June 22, 2022. Retrieved June 24, 2022.
 "Ms. Marvel Production Notes" (PDF). Disney Media and Entertainment Distribution. Archived (PDF) from the original on June 18, 2022. Retrieved June 17, 2022.
 Wasserman, Ben (December 15, 2021). "Ms. Marvel Star Addresses Their Hopes for the Show and Its Fans". Comic Book Resources. Archived from the original on March 18, 2022. Retrieved April 5, 2022.
 Scott, Ryan (March 15, 2022). "Ms. Marvel Trailer Breakdown: Kamala Khan Enters The MCU With Brand New Powers". /Film. Archived from the original on March 15, 2022. Retrieved March 16, 2022.
 Grauso, Alisha (June 23, 2022). "Ms. Marvel Cast Guide: Every Marvel Character". Screen Rant. Archived from the original on June 23, 2022. Retrieved June 23, 2022.
 Silliman, Brian (June 8, 2022). "'Ms. Marvel' Series Premiere Takes Us Into the MCU's Meta-Fandom as a New Hero Is Born". SyFy Wire. Archived from the original on June 8, 2022. Retrieved June 8, 2022.
 Amin, Arezou (June 29, 2022). "'Ms. Marvel' Episode 4 Review: Chaos and ClanDestines Come to Karachi". Collider. Archived from the original on June 29, 2022.
 Khan, Rabab (May 31, 2022). "Ms. Marvel Character Posters Reveal New Look At The Main Cast". Game Rant. Archived from the original on May 31, 2022. Retrieved June 4, 2022.
 Oddo, Marco Vito (June 22, 2022). "'Ms. Marvel' Recycles an Obscure Comics Supervillain Group — and Makes It Better". Collider. Archived from the original on June 23, 2022. Retrieved June 23, 2022.
 Tikader, Agrima (June 8, 2022). "Iman Vellani shines as Ms Marvel in Episode 1 of the Disney+Hotstar show". The Telegraph. Archived from the original on June 9, 2022. Retrieved June 10, 2022.
 Adlakha, Siddhant (June 8, 2022). "Ms. Marvel Series-Premiere Recap: Diary of a Wimpy Teen". Vulture. Archived from the original on June 8, 2022. Retrieved June 10, 2022.
 Ahmed, Musanna (February 19, 2021). ""Trying To Understand My Character's Journey Was Difficult": Interview with Alyy Khan, Star of Mogul & The Serpent". Film Inquiry. Archived from the original on February 19, 2021. Retrieved April 11, 2021.
 Brail, Nathaniel (June 8, 2022). "Ms. Marvel Features Surprising Cameo Only True Marvel Fans Will Recognize". ComicBook.com. Archived from the original on June 9, 2022. Retrieved June 9, 2022.
 Adams, Timothy (June 21, 2022). "Ms. Marvel Episode 3 Clip Features a Bollywood-Style Wedding Dance". ComicBook.com. Archived from the original on June 23, 2022. Retrieved June 23, 2022.
 "Disney+ Debuts Marvel Studios' "A Fan's Guide To Ms. Marvel"". Disney Media and Entertainment Distribution. June 1, 2022. Archived from the original on June 2, 2022. Retrieved June 1, 2022.
 "Ms. Marvel - WGA Directory". Writers Guild of America West. June 2, 2022. Archived from the original on June 1, 2022. Retrieved June 3, 2022.
 Roberts, Kayleigh (May 16, 2022). "Here's When Every Episode of 'Ms. Marvel' Comes Out". Cosmopolitan. Archived from the original on May 17, 2022. Retrieved May 16, 2022.
 Couto, Anthony (September 6, 2016). "Quesada on Marvel's Diverse Audience, Ms. Marvel's Future in TV & Film". Comic Book Resources. Archived from the original on September 9, 2016. Retrieved July 31, 2017.
 Osborn, Alex (May 12, 2018). "Feige: MCU Has 'Plans' to Introduce Ms. Marvel After Captain Marvel". IGN. Archived from the original on May 13, 2018. Retrieved April 8, 2019.
 Coggan, Devan (May 11, 2022). "'Ms. Marvel' co-creator Sana Amanat on bringing Kamala Khan from page to screen". Entertainment Weekly. Archived from the original on May 11, 2022. Retrieved May 11, 2022.
 Kit, Borys; Goldberg, Lesley (August 23, 2019). "'Ms. Marvel' Series in the Works for Disney+ (Exclusive)". The Hollywood Reporter. Archived from the original on August 23, 2019. Retrieved August 23, 2019.
 Mathai, Jeremy (June 7, 2022). "Ms. Marvel Head Writer Bisha K. Ali On Her Shah Rukh Khan Obsession And Avoiding Clichés [Interview]". /Film. Archived from the original on June 8, 2022. Retrieved June 12, 2022.
 Davids, Brian (June 15, 2022). "How Bisha K. Ali Made Her Own Luck with 'Ms. Marvel'". The Hollywood Reporter. Archived from the original on June 15, 2022. Retrieved June 16, 2022.
 Dominguez, Noah (August 23, 2019). "Ms. Marvel Series Confirmed, Character Will Appear In Movies". Comic Book Resources. Archived from the original on August 24, 2019. Retrieved August 24, 2019.
 Couch, Aaron (September 18, 2020). "'Ms. Marvel' Finds Directors in Pakistani Oscar Winner, 'Bad Boys For Life' Filmmakers (Exclusive)". The Hollywood Reporter. Archived from the original on September 18, 2020. Retrieved September 18, 2020.
 Alber, Caitlin (May 15, 2021). "Here's Who's Directing Marvel's 'Secret Invasion' Disney+ Series". Collider. Archived from the original on May 15, 2021. Retrieved May 15, 2021.
 Kit, Borys (May 19, 2021). "'Bad Boys for Life' Filmmakers to Direct 'Batgirl' Movie for Warner Bros. (Exclusive)". The Hollywood Reporter. Archived from the original on May 19, 2021. Retrieved May 19, 2021.
 Weiss, Josh (December 11, 2020). "Embiggen! Ms. Marvel Series Coming to Disney+, Drops Sizzle Reel With First Footage". Syfy Wire. Archived from the original on December 12, 2020. Retrieved December 12, 2020.
 Salazar, Andrew J. (September 6, 2019). "'What If...?' Showrunner Ashley Bradley Talks 'Trollhunters', The Beginning Of 'What If...?', Captain Carter, The Watcher & Star-Lord T'Challa (Exclusive)". Discussing Film. Archived from the original on September 22, 2019. Retrieved September 27, 2019.
 "What If..? Production Brief" (PDF). Disney Media and Entertainment Distribution. July 30, 2021. Archived (PDF) from the original on August 1, 2021. Retrieved August 1, 2021.
 "Moon Knight - WGA Directory". Writers Guild of America West. February 23, 2022. Archived from the original on February 23, 2022. Retrieved February 23, 2022.
 Radish, Christina (June 9, 2022). "'Ms. Marvel's Iman Vellani and Creatives Discuss Bringing Kamala Khan to Screen and AvengersCon". Collider. Archived from the original on June 10, 2022. Retrieved June 12, 2022.
 Travis, Ben (May 9, 2022). "Ms. Marvel's Powers Are 'Not An Exact Translation' From The Comics, Says Kevin Feige – Exclusive Image". Empire. Archived from the original on May 9, 2022. Retrieved May 9, 2022.
 Wittmer, Carrie (June 7, 2022). "The Dawn Of Iman Vellani". Elite Daily. Archived from the original on June 11, 2022. Retrieved June 15, 2022.
 Anderson, Jenna (November 19, 2020). "Ms. Marvel: Disney+ Series Casts Matt Lintz in Key Role". ComicBook.com. Archived from the original on November 20, 2020. Retrieved November 20, 2020.
 Hermanns, Grant (December 10, 2020). "Iman Vellani is Kamala Khan in First Ms. Marvel Sizzle Reel!". ComingSoon.net. Archived from the original on December 11, 2020. Retrieved December 12, 2020.
 Martinez, Kiko (December 12, 2020). "San Antonio native Saagar Shaikh lands acting role in upcoming Disney+ series Ms. Marvel". San Antonio Current. Archived from the original on December 12, 2020. Retrieved December 12, 2020.
 Barnhardt, Adam (December 11, 2020). "Ms. Marvel Actor Teases Arrival of Inhumans". ComicBook.com. Archived from the original on December 12, 2020. Retrieved December 14, 2020.
 Kroll, Justin (April 28, 2021). "Rish Shah To Co-Star In The Netflix Pic 'Strangers' From 'Someone Great' Director". Deadline Hollywood. Archived from the original on April 28, 2021. Retrieved May 2, 2021.
 Couch, Aaron (February 19, 2021). "'Ms. Marvel' Enlists Laurel Marsden (Exclusive)". The Hollywood Reporter. Archived from the original on February 19, 2021. Retrieved February 19, 2021.
 Nakli, Laith [@laithnakli] (February 28, 2021). "And that's a wrap for #msmarvel what an experience. Just when I thought it couldn't get better it did when @alysiareiner joined the cast. 16 years of friendship and working together. Thank you @minib622 and #badass directors @fallahbilall @adilelarbi @meeraonthewall and the incredible cast!! Can't wait for the world to see this show. Also can't wait to see you New York !! #msmarvel #marvel #atl #representation #muslimsontv". Retrieved February 28, 2021 – via Instagram.
 "Fawad Khan confirms role in Ms Marvel, says he had 'good fun' shooting for it". Hindustan Times. December 16, 2021. Archived from the original on December 16, 2021. Retrieved December 17, 2021.
 "Mehwish Hayat joins Fawad Khan, Nimra Bucha in 'Ms Marvel'". The Express Tribune. January 14, 2022. Archived from the original on January 14, 2022. Retrieved January 14, 2022.
 "Today's the Day! Catch Adaku Ononogbo starring in Ms Marvel as Fariha. The wait is finally over! The trailer is out for Ms. Marvel, an Original series from Marvel Studios, starts streaming June 8 on @disneyplus . #MsMarvel. Set your reminders for June 8 on Disney Plus !!". International Artists Management. March 15, 2022. Archived from the original on May 3, 2022. Retrieved May 2, 2022.
 Ridgely, Charlie (May 19, 2022). "Ms. Marvel Brings Back Spider-Man: No Way Home, Inventing Anna Actor". ComicBook.com. Archived from the original on May 20, 2022. Retrieved May 20, 2022. – via Project Profile (archive)
 Tartaglione, Nancy (May 6, 2022). "Farhan Akhtar To Appear In Disney Plus Series 'Ms. Marvel'". Deadline Hollywood. Archived from the original on May 6, 2022. Retrieved May 6, 2022.
 Cordero, Rosy (March 14, 2022). "Anjali Bhimani Boards 'Ms. Marvel' In A Recurring Role". Deadline Hollywood. Archived from the original on March 14, 2022. Retrieved March 14, 2022.
 O'Hara, Helen (Summer 2022). "Punching Through". Empire. No. 402. pp. 84–89.
 Robinson, Tasha (June 9, 2022). "Ms. Marvel's directors say episode 1's biggest action sequence wasn't in the script". Polygon. Archived from the original on June 9, 2022. Retrieved June 11, 2022.
 Sharma, Dishya (January 16, 2021). "Exclusive: Loki star Tom Hiddleston cycled past Mohan Kapur but Ms Marvel star didn't spot him for This reason". Pinkvilla. Archived from the original on January 17, 2021. Retrieved January 17, 2021.
 Kit, Borys; Couch, Aaron (November 20, 2020). "Marvel's 'Black Panther' Sequel Shoot to Begin in July (Exclusive)". The Hollywood Reporter. Archived from the original on November 20, 2020. Retrieved November 20, 2020.
 White, Brett (June 9, 2022). "'Ms. Marvel' Filming Locations: Was the New Disney+ Show Filmed in Jersey City?". Decider. Archived from the original on June 9, 2022. Retrieved June 9, 2022.
 Hickson, Colin (June 29, 2020). "Ms. Marvel Working Title Seemingly Confirms the Show's Setting". Comic Book Resources. Archived from the original on September 10, 2020. Retrieved November 11, 2020.
 Davids, Brian (June 13, 2022). "'Ms. Marvel' Directors Adil and Bilall Discuss Kevin Feige's 'Batgirl' Fandom and Michael Keaton's Batsuit Mobility". The Hollywood Reporter. Archived from the original on June 14, 2022. Retrieved June 14, 2022.
 Davids, Brian (June 20, 2022). "'Ms. Marvel' Director Meera Menon Discusses How She Relates to Kamala Khan and Breaks Down Episode Two's Training Montage". The Hollywood Reporter. Archived from the original on June 22, 2022. Retrieved June 21, 2022.
 O'Loughlin, Jules [@jules.oloughlin] (March 23, 2021). "#marvel". Retrieved April 22, 2021 – via Instagram.
 Robinson, Tasha (June 12, 2022). "Ms. Marvel's directors wanted their show to be 'Into the Spider-Verse in live action'". Polygon. Archived from the original on June 13, 2022. Retrieved June 13, 2022.
 Arbi, Adil El [@adilelarbi] (March 5, 2021). "Jumu3a Mubarak!!! Last shooting day for us on Ms. Marvel. Working with legendary producer Kevin Feige has been a blast! Merci frerooo!Shokran to the producers Sana Amanat & Jenna Berger, the writers, the whole crew and cast! You are family FOR LIFE! Stay tuned for the end of year insha Allah. Good luck to @sharmeenobaidchinoy with her episodes Al Hamdoulillah ⚡⚡⚡ #msmarvel #drarriesunited 🇲🇦". Retrieved March 5, 2021 – via Instagram.
 Kuperinsky, Amy (May 4, 2021). "First look at N.J.'s Ms. Marvel, 'The Marvels' movie teaser". NJ.com. Archived from the original on May 4, 2021. Retrieved May 6, 2021.
 Hood, Cooper (May 1, 2021). "Ms. Marvel Set Photos Reveal First Look at Kamala Khan's Comics Accurate Costume". Screen Rant. Archived from the original on May 2, 2021. Retrieved May 2, 2021.
 Frater, Patrick (May 11, 2021). "'Ms. Marvel' Series Completes Production in Thailand Despite Virus Resurgence (Exclusive)". Variety. Archived from the original on May 11, 2021. Retrieved May 11, 2021.
 Dominguez, Noah (January 26, 2022). "Ms. Marvel Set Photos Confirm Reshoots Are Underway". Comic Book Resources. Archived from the original on January 30, 2022. Retrieved February 1, 2022.
 Frei, Vincent (May 25, 2022). "Ms. Marvel". Art of VFX. Archived from the original on June 4, 2022. Retrieved June 3, 2022.
 Tangcay, Jazz (May 16, 2022). "'The Marvels' Composer Laura Karpman Scores 'Ms. Marvel' on Disney+ (Exclusive)". Variety. Archived from the original on May 16, 2022. Retrieved May 16, 2022.
 "'Ms. Marvel' Vol. 1 (Episodes 1-3) Soundtrack Album Details". Film Music Reporter. June 21, 2022. Archived from the original on June 21, 2022. Retrieved June 22, 2022.
 "First Track from 'Ms. Marvel' Soundtrack Released". Film Music Reporter. June 7, 2022. Archived from the original on June 8, 2022. Retrieved June 8, 2022.
 D'Alessandro, Anthony (November 12, 2021). "Marvel's 'Moon Knight', 'She-Hulk', 'Ms. Marvel' & 'Hawkeye' Footage Unveiled In Disney+ Day Special; New Series Set". Deadline Hollywood. Archived from the original on November 13, 2021. Retrieved November 12, 2021.
 Skrebels, Joe (March 15, 2022). "Ms. Marvel: First Trailer Reveals June Release Date". IGN. Archived from the original on March 15, 2022. Retrieved March 15, 2022.
 Pulliam-Moore, Charles (March 15, 2022). "Ms. Marvel's first trailer teases a cosmic take on Kamala Khan". The Verge. Archived from the original on March 15, 2022. Retrieved March 15, 2022.
 Sanchez, Gabrielle (March 15, 2022). "Kamala Khan makes her debut in the trailer for Ms. Marvel". The A.V. Club. Archived from the original on March 15, 2022. Retrieved March 15, 2022.
 Ryan, Danielle (March 24, 2022). "The MCU's Ms. Marvel Has Very Different Powers Than The Comic Version". /Film. Archived from the original on March 25, 2022. Retrieved March 25, 2022.
 Codega, Linda (March 15, 2022). "Ms. Marvel's First Trailer Brings a Radical New Kamala Khan to the MCU". Gizmodo. Archived from the original on March 15, 2022. Retrieved March 15, 2022.
 Mboho, Edidiong (March 15, 2022). "'Ms. Marvel' Poster Calls Upon Comic Book Nostalgia". Collider. Archived from the original on March 15, 2022. Retrieved March 16, 2022.
 Ridgely, Charlie (May 23, 2022). "Disney+: Every Movie & TV Show Arriving in June 2022". ComicBook.com. Archived from the original on May 23, 2022. Retrieved May 24, 2022.
 Paige, Rachel (January 12, 2021). "Introducing Marvel Must Haves". Marvel.com. Archived from the original on January 20, 2021. Retrieved January 31, 2021.
 Paige, Rachel (June 7, 2022). "Shop Marvel Must Haves: AvengerCon T-Shirts". Marvel.com. Archived from the original on June 8, 2022. Retrieved June 8, 2022.
 Perine, Aaron (June 8, 2022). "Ms. Marvel Features Another Hidden QR Code For A Free Comic". ComicBook.com. Archived from the original on June 8, 2022. Retrieved June 8, 2022.
 "Free Ms. Marvel Comics Presented by Marvel Unlimited". Marvel.com. June 8, 2022. Retrieved June 14, 2022.
 Odman, Sydney (June 3, 2022). "'Ms. Marvel' Team Talks Matching Cast Tattoos, Muslim Superhero Representation". The Hollywood Reporter. Archived from the original on June 4, 2022. Retrieved June 3, 2022.
 Couch, Aaron (July 29, 2021). "'Hawkeye' Sets November Release Date on Disney+". The Hollywood Reporter. Archived from the original on July 29, 2021. Retrieved July 29, 2021.
 Mitovich, Matt Webb (August 2, 2021). "Matt's Inside Line: Scoop on Riverdale, #OneChicago, SEAL Team, Titans, Stargirl, Lucifer, Good Trouble and More!". TVLine. Archived from the original on August 2, 2021. Retrieved August 2, 2021.
 Vary, Adam B. [@adambvary] (September 22, 2021). "Update: I've confirmed what was obvious to most Marvel Studios observers: #MsMarvel has moved from a late 2021 release on Disney Plus to 2022" (Tweet). Archived from the original on September 23, 2021. Retrieved September 23, 2021 – via Twitter.
 Brown, Tracy (August 23, 2019). "Here are all the Marvel shows coming to Disney+, from 'Ms. Marvel' to 'Moon Knight'". Los Angeles Times. Archived from the original on August 24, 2019. Retrieved August 24, 2019.
 "'Ms Marvel' to screen in theatres across Pakistan". The Express Tribune. May 3, 2022. Archived from the original on May 3, 2022. Retrieved May 3, 2022.
 Oddo, Marco Vitto (May 3, 2022). "'Ms. Marvel': Full Series to Get Theatrical Release in Pakistan". Collider. Archived from the original on May 3, 2022. Retrieved May 9, 2022.
 "Ms. Marvel: Season 1". Rotten Tomatoes. Fandango Media. Retrieved June 29, 2022.
 "Ms. Marvel: Season 1". Metacritic. Red Ventures. Retrieved June 11, 2022.
 Han, Angie (June 7, 2022). "Disney+'s 'Ms. Marvel': TV Review". The Hollywood Reporter. Archived from the original on June 7, 2022. Retrieved June 7, 2022.
 Jackson, Destiny (June 7, 2022). "Ms. Marvel Review". Empire. Archived from the original on June 7, 2022. Retrieved June 7, 2022.
 Sepinwall, Alan (June 7, 2022). "As a Coming-of-Age Tale, 'Ms. Marvel' Kicks Ass. As a Superhero Origin Story, Maybe Not". Rolling Stone. Archived from the original on June 7, 2022. Retrieved June 7, 2022.
 Framke, Caroline (June 7, 2022). "'Ms. Marvel' Brings 'Into the Spider-Verse' Energy to Charming, Refreshing New Heroine: TV Review". Variety. Archived from the original on June 7, 2022. Retrieved June 7, 2022.
 "Malala Yousafzai Gives a Shout Out to Ms Marvel for Reflecting 'The Lives of a Pakistani Immigrant Family'". News18. June 8, 2022. Archived from the original on June 8, 2022. Retrieved June 8, 2022.
 Malala (June 7, 2022). "Ms. Marvel creator shares the "crazy, surreal, emotional" journey of bringing Kamala Khan to life". Archived from the original on June 8, 2022. Retrieved June 7, 2022 – via Bulletin.
 Fraser, Emma (June 7, 2022). "Ms. Marvel: Series Premiere Review". IGN. Archived from the original on June 7, 2022. Retrieved June 7, 2022.
 Porter, Kathryn (June 7, 2022). "Ms. Marvel Is the Boost the MCU Needs, but Not the Kamala Story Hardcore Fans Might Want". Paste. Archived from the original on June 7, 2022. Retrieved June 7, 2022.
 Zaheer, Mohammad (June 7, 2022). "Why Marvel has struck gold with Muslim superhero Ms Marvel". BBC Culture. Archived from the original on June 8, 2022. Retrieved June 7, 2022.
 Francisco, Eric (June 7, 2022). "'Ms. Marvel' review: A radiant blast of pure, unbridled joy for the MCU". Inverse. Archived from the original on June 7, 2022. Retrieved June 7, 2022.
 Pulliam-Moore, Charles (June 7, 2022). "Ms. Marvel feels like the future of the MCU". The Verge. Archived from the original on June 7, 2022. Retrieved June 7, 2022.
 Shaffi, Sarah (June 7, 2022). "'Ms. Marvel' first look review: Muslim superhero adds new magic to the MCU". NME. Archived from the original on June 7, 2022. Retrieved June 7, 2022.
 Hadadi, Roxana (June 7, 2022). "Ms. Marvel's Enthusiasm Is Infectious". Vulture. Archived from the original on June 7, 2022. Retrieved June 7, 2022.
 Dodge, John (June 8, 2022). "Ms. Marvel Is the MCU's Latest Series to Suffer Review-Bombing". Comic Book Resources. Archived from the original on June 9, 2022. Retrieved June 12, 2020.
 Bentz, Adam (June 8, 2022). "Ms. Marvel Review Bombed On IMDb After Disney+ Release". Screen Rant. Archived from the original on June 8, 2022. Retrieved June 12, 2020.
 Couch, Aaron (November 10, 2019). "Marvel's Kevin Feige Breaks Silence on Scorsese Attack: "It's Unfortunate" (Exclusive)". The Hollywood Reporter. Archived from the original on November 10, 2019. Retrieved November 10, 2019.
 Aguilar, Matthew (February 14, 2019). "'Captain Marvel's Brie Larson Wants Ms. Marvel In The Sequel". ComicBook.com. Archived from the original on August 18, 2019. Retrieved January 24, 2020.
 Bui, Hoai-Tran (December 10, 2020). "'Ms. Marvel' First Look Teases New Disney+ Series, Character Confirmed to Appear in 'Captain Marvel 2'". /Film. Archived from the original on December 10, 2020. Retrieved December 10, 2020.
 Couch, Aaron (May 3, 2021). "Marvel Unveils 'Black Panther II' Title, First 'Eternals' Footage and More". The Hollywood Reporter. Archived from the original on May 3, 2021. Retrieved May 3, 2021.
 D'Alessandro, Anthony (February 24, 2021). "'WandaVision': Elizabeth Olsen Character Will Head To 'Doctor Strange 2' Pic; No Season 2 Planned Yet". Deadline Hollywood. Archived from the original on February 24, 2021. Retrieved February 24, 2021.
 Hussaini, Syed Fahadullah (September 7, 2021). "Ms. Marvel Star Reveals Show Characters Return In Captain Marvel 2". Screen Rant. Archived from the original on September 7, 2021. Retrieved September 7, 2021.
External links
Ms. Marvel
at Wikipedia's sister projects
Media from Commons
Quotations from Wikiquote
Data from Wikidata
Official website Edit this at Wikidata at Marvel.com
Ms. Marvel at IMDb Edit this at Wikidata
Ms. Marvel on Disney+ Edit this at Wikidata
Ms. Marvel
Portals:
icon Television
flag United States
icon Speculative fiction
icon Comics
Categories: 2020s American comedy-drama television series2020s American science fiction television series2020s American superhero comedy television series2020s American teen drama television series2022 American television series debutsAmerican action adventure television seriesAsian-American televisionComing-of-age television showsDisney+ original programmingEnglish-language television showsMarvel Cinematic Universe: Phase Four television seriesMs. Marvel (TV series)Pakistani-American culturePartition of India in fictionSerial drama television seriesSuperheroine television showsTeen superhero television seriesTelevision series about teenagersTelevision series by Marvel StudiosTelevision series set in the 2020sTelevision shows filmed at Trilith StudiosTelevision shows filmed in AtlantaTelevision shows filmed in Georgia (U.S. state)Television shows filmed in New JerseyTelevision shows filmed in ThailandTelevision shows set in KarachiTelevision shows set in New JerseyTelevision shows set in the British Raj
Navigation menu
Not logged in
Talk
Contributions
Create account
Log in
ArticleTalk
ReadEditView history
Search
Search Wikipedia
Main page
Contents
Current events
Random article
About Wikipedia
Contact us
Donate
Contribute
Help
Learn to edit
Community portal
Recent changes
Upload file
Tools
What links here
Related changes
Special pages
Permanent link
Page information
Cite this page
Wikidata item
Print/export
Download as PDF
Printable version
In other projects
Wikiquote

Languages
العربية
Español
فارسی
Français
Bahasa Indonesia
ਪੰਜਾਬੀ
Português
Русский
中文
22 more
Edit links
This page was last edited on 2 July 2022, at 11:41 (UTC).
Text is available under the Creative Commons Attribution-ShareAlike License 3.0; additional terms may apply. By using this site, you agree to the Terms of Use and Privacy Policy. Wikipedia® is a registered trademark of the Wikimedia Foundation, Inc., a non-profit organization.
Privacy policyAbout WikipediaDisclaimersContact WikipediaMobile viewDevelopersStatisticsCookie statementWikimedia FoundationPowered by MediaWiki
"""
```
